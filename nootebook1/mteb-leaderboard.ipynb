## Étape 2 : Définir les modèles
Nous allons définir les modèles sur lesquels nous souhaitons effectuer des tests de performance.

model_names : liste des modèles d’intégration à évaluer.

Inclut des modèles de HuggingFace (par exemple, MiniLM, MPNet, BGE, Jina embeddings).

Tous seront testés sur la même tâche de référence.

```python
model_names = [
    "sentence-transformers/all-MiniLM-L6-v2",
    "sentence-transformers/all-mpnet-base-v2",
    "sentence-transformers/all-distilroberta-v1",
    "sentence-transformers/paraphrase-MiniLM-L6-v2",
    "sentence-transformers/paraphrase-mpnet-base-v2",
    "intfloat/e5-base-v2",
    "BAAI/bge-base-en-v1.5",
    "jinaai/jina-embeddings-v2-base-en",
]
```

## Étape 3 : Charger les modèles et exécuter le test de performance

SentenceTransformer(nom_du_modèle) : charge le modèle pré-entraîné.

MTEB(tâches=tâches) : prépare l'évaluation sur STSBenchmark (Similarité textuelle sémantique).

evaluation.run(...): exécute l'analyse comparative et enregistre les résultats dans results_local_comp/... avec un sous-dossier distinct par modèle.

model_name.replace('/', '_')  garantit l'absence de noms de dossiers invalides.

verbosité=0 : réduit la sortie au minimum.

```python
from sentence_transformers import SentenceTransformer
import mteb
from mteb import MTEB

tasks = mteb.get_tasks(tasks=["STSBenchmark"], languages=["eng"])
results_base_dir = "results_local_comp"

for model_name in model_names:
    print(f"Evaluating {model_name} ...")
    model = SentenceTransformer(model_name)
    evaluation = MTEB(tasks=tasks)
    evaluation.run(
        model,
        output_folder=f"{results_base_dir}/{model_name.replace('/', '_')}",
        verbosity=0,
    )
print("Benchmarking done.")
```

## Étape 4 : Configuration de l’extraction des résultats

Nous allons configurer le bloc d'extraction des résultats.

Utilitaires d'importation pour la gestion des fichiers (os), des résultats JSON (json) et des tables (pandas).

results_base_dir a été mis à jour avec le chemin absolu où les résultats ont été enregistrés.

```python
import os
import json
import pandas as pd

results_base_dir = "/content/results_local_comp"
results_base_dir = "/kaggle/working/results_local_comp"
```

## Étape 5 : Définir une fonction auxiliaire pour lire les métriques
Nous allons définir la fonction auxiliaire,

La fonction démarre dans un dossier modèle donné et navigue sur deux niveaux de sous-dossiers.

Il recherche ensuite un fichier nommé STSBenchmark.json dans le dossier de second niveau.

Si le fichier existe, il charge les données JSON.

À partir des données, il extrait les résultats des tests (sous « scores » : « test »).

Métrique : nom de la métrique (ex. : précision, Pearson, Spearman).

Valeur : valeur métrique (formatée à 4 décimales).

Modèle : le nom du dossier du modèle.

Si une étape échoue (dossiers manquants, fichier manquant ou absence de scores), elle affiche un message clair et renvoie None.

```python
def extract_metrics(model_folder):
    try:
        first_level = next(os.walk(model_folder))[1]
        if not first_level:
            print(f"No first level folders in {model_folder}")
            return None
        first_level_folder = first_level[0]

        second_level_path = os.path.join(model_folder, first_level_folder)
        second_level_walk = next(os.walk(second_level_path))
        second_level = second_level_walk[1]
        if not second_level:
            print(f"No second level folders in {second_level_path}")
            return None
        second_level_folder = second_level[0]

        json_path = os.path.join(
            second_level_path, second_level_folder, "STSBenchmark.json")
        if not os.path.isfile(json_path):
            print(f"No STSBenchmark.json file at {json_path}")
            return None

        with open(json_path, "r") as f:
            result_data = json.load(f)

        scores = result_data.get("scores", {})
        test_scores = scores.get("test", [])

        if not test_scores:
            print(f"No test scores found in {json_path}")
            return None

        metrics = test_scores[0]
        df = pd.DataFrame(list(metrics.items()), columns=["Metric", "Value"])
        df["Value"] = df["Value"].apply(
            lambda x: f"{x:.4f}" if isinstance(x, float) else x)
        df["Model"] = os.path.basename(model_folder)
        return df
    except Exception as e:
        print(f"Error processing folder {model_folder}: {e}")
        return None
```

## Étape 6 : Collecte des résultats pour tous les modèles

Nous allons recueillir les résultats pour tous les modèles.

Liste de tous les dossiers de résultats de modèles dans results_local_comp.

Parcourt chaque dossier de modèle.

Extrait les métriques à l'aide de extract_metrics.

Stocke les résultats dans une liste de DataFrames.

```python
model_folders = [os.path.join(results_base_dir, d) for d in os.listdir(
    results_base_dir) if os.path.isdir(os.path.join(results_base_dir, d))]

dfs = []
for folder in model_folders:
    df_metrics = extract_metrics(folder)
    if df_metrics is not None:
        dfs.append(df_metrics)
```

## Étape 7 : Combiner et afficher toutes les métriques

Nous afficherons les résultats :

Concaténe tous les DataFrames en un seul grand tableau (tous les modèles + scores).

Imprime les résultats de l'évaluation finale.

```python
if dfs:
    combined_df = pd.concat(dfs).reset_index(drop=True)
    print(combined_df)
else:
    print("No metrics found for any models.")

display(combined_df)
```

## Étape 8 : Visualiser le résultat
Nous allons représenter graphiquement les résultats et les indicateurs utilisés sont :

cosine_pearson :  Corrélation de Pearson entre les scores de similarité cosinus du modèle et les jugements humains.

cosine_spearman :  Corrélation de Spearman (ordre de classement) pour les scores de similarité cosinus par rapport aux classements humains.

euclidean_pearson :  Corrélation de Pearson pour les distances euclidiennes converties en similarité, comparée aux scores humains.

euclidean_spearman :  Corrélation de Spearman pour les similarités basées sur la distance euclidienne par rapport aux classements humains.

manhattan_pearson :  Corrélation de Pearson pour les similarités basées sur la distance de Manhattan (L1).

manhattan_spearman :  Corrélation de Spearman pour les similarités basées sur la distance de Manhattan.

pearson :  Corrélation de Pearson sur les scores de similarité principaux du modèle (souvent le cosinus).

Spearman :  Corrélation de rang de Spearman sur les scores de similarité primaires.

main_score :  Un score global ou une mesure de synthèse des performances du modèle (peut être une mesure moyenne ou principale utilisée pour le classement).

Langues :  Nombre de langues couvertes par l'évaluation.

hf_subset :  Indique un sous-ensemble spécifique de données d’évaluation (souvent issues des ensembles de données HuggingFace).

```python
import matplotlib.pyplot as plt
import pandas as pd

combined_df['Value'] = pd.to_numeric(combined_df['Value'], errors='coerce')

pivot_df = combined_df.pivot(index='Model', columns='Metric', values='Value')

pivot_df.plot(kind='bar', figsize=(12, 6))

plt.title('Model Performance on STSBenchmark Metrics')
plt.ylabel('Metric Value')
plt.xlabel('Model')
plt.xticks(rotation=45, ha='right')

plt.legend(title='Metric', bbox_to_anchor=(
    1.02, 1), loc='upper left', borderaxespad=0)
plt.tight_layout()
plt.show()
```
