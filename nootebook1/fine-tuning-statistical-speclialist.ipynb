# Fine tuning

```python
import keras
import keras_hub
import numpy as np

print("\nLoading Qwen3 0.6B model...")
qwen3_lm = keras_hub.models.Qwen3CausalLM.from_preset("kaggle://keras/qwen-3/keras/qwen3_0.6b_en/1")

# Prepare training data
training_texts = [
    "You are statistical speclialist",
    "Statistics is a data analysis tool.",
    "In statistics, mean is calculated by adding up all the values in a dataset and dividing the sum by the number of values.",

]

# Compile for training
qwen3_lm.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(1e-4),
    metrics=["accuracy"]
)

# Fine-tune the model
qwen3_lm.fit(x=training_texts, batch_size=2, epochs=10)

# Generate with fine-tuned model
response = qwen3_lm.generate("what is standard deviation", max_length=50)
print(response)
```

```python
# Batch generation with multiple prompts
prompts = ["what is standard deviation",
           "nonparametric statistical method"
          ]
responses = qwen3_lm.generate(prompts, max_length=30)
for prompt, response in zip(prompts, responses):
    print(f"Prompt: {prompt}")
    print(f"Response: {response}\n")
```
